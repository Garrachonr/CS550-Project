```python
import pandas as pd
import krippendorff

#Load Excel Data
file_path = 'conversations.xlsx'
data = pd.read_excel(file_path, sheet_name=0)

#Clean data
data_clean = data.drop(columns=['Unnamed: 0'])
data_np = data_clean.to_numpy().T

alpha = krippendorff.alpha(reliability_data=data_np)

print(f"Krippendorff's Alpha: {alpha}")
```
