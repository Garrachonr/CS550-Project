```python
import numpy as np
from scipy.stats import pearsonr

resultados_evaluacion_manual = [0.6984, 0.8051, 0.8294, 0.9615]
resultados_evaluacion_determinista = [0.8510, 0.8615, 0.8592, 0.8744]

#Compute the pearson coeficient
coeficiente_pearson, p_valor = pearsonr(resultados_evaluacion_manual, resultados_evaluacion_determinista)

print(f"Coeficiente de correlación de Pearson: {coeficiente_pearson}")
print(f"P-valor: {p_valor}")
```
